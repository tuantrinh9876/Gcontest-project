```python
sale_type = df.groupby('Hình thức bán hàng ')['Doanh thu thuần'].sum()
sale_type
```

```python
# Group by area and year
revenue_by_area = df.groupby(['NĂM ','KHU VỰC '])['Doanh thu thuần'].sum().unstack()
revenue_by_area
```

```python
# Create the stacked bar chart
revenue_by_area.plot(kind='bar', stacked=True)
plt.title('Revenue by area and by year')
plt.xlabel('Year')
plt.ylabel('Net Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by area and year
gross_profit_by_area = df.groupby(['NĂM ','KHU VỰC '])['Lợi nhuận gộp'].sum().unstack()
gross_profit_by_area
```

```python
# Create the stacked bar chart
gross_profit_by_area.plot(kind='bar', stacked=True)
plt.title('Gross profit by area and by year')
plt.xlabel('Year')
plt.ylabel('Net Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
net_profit_by_area = df.groupby(['NĂM ','KHU VỰC '])['Lợi nhuận NET'].sum().unstack()
net_profit_by_area
```

```python
# Create the stacked bar chart
net_profit_by_area.plot(kind='bar', stacked=True)
plt.title('Gross profit by area and by year')
plt.xlabel('Year')
plt.ylabel('Net Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
# Doanh thu trung bình mỗi chi nhánh 
avg_revenue = df.groupby(['KHU VỰC ','NĂM '])['Doanh thu thuần'].mean().unstack()
avg_revenue
```

```python
# Lợi nhuận trung bình mỗi chi nhánh
avg_net_profit = df.groupby(['KHU VỰC ','NĂM '])['Lợi nhuận NET'].mean().unstack()
avg_net_profit
```

```python
# %chi phí bỏ ra trên doanh thu
avg_cost_percent = df.groupby(['KHU VỰC ','NĂM '])['Cost percent'].mean().unstack()
avg_cost_percent
```

```python
# Tính tổng Doanh thu thuần và Kế hoạch doanh thu thuần theo khu vực
KPT_gross_revenue = df.groupby("KHU VỰC ")[["Doanh thu thuần", "Kế hoạch DTT"]].sum().reset_index()

# Thêm cột tỷ lệ hoàn thành kế hoạch
KPT_gross_revenue["Tỷ lệ hoàn thành"] = (KPT_gross_revenue["Doanh thu thuần"] / KPT_gross_revenue["Kế hoạch DTT"]) * 100
KPT_gross_revenue
```

```python
# Tính tổng Lợi nhuận Net và Kế hoạch lợi nhuận net theo khu vực
KPT_net_profit = df.groupby("KHU VỰC ")[["Lợi nhuận NET", "Kế hoạch LNN"]].sum().reset_index()

# Thêm cột tỷ lệ hoàn thành kế hoạch
KPT_net_profit["Tỷ lệ hoàn thành"] = (KPT_net_profit["Lợi nhuận NET"] / KPT_net_profit["Kế hoạch LNN"]) * 100
KPT_net_profit
```
